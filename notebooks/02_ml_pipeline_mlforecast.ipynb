# 02 - MLForecast Features and Pipeline
This notebook exercises the reusable pipeline and core `mlforecast` capabilities in this project.

```
from mlforecast_realworld.ml.pipeline import ForecastPipeline
pipeline = ForecastPipeline()
df = pipeline.prepare_training_data(download=False)
```

```
fitted_values = pipeline.fit(df)
fitted_values.head()
```

```
cv_df, cv_summary = pipeline.cross_validate(df)
cv_summary
```

```
preds = pipeline.forecast(horizon=14)
preds.head()
```

```
# Optional LightGBM CV demo
lightgbm_preds = pipeline.run_lightgbm_cv(df)
lightgbm_preds.head()
```
