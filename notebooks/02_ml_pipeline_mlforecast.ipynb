# 02 - ML Pipeline, Method Testing, and Accuracy Report
This notebook runs the reusable production pipeline and reports cross-validation accuracy per model.

```
from mlforecast_realworld.ml.pipeline import ForecastPipeline
pipeline = ForecastPipeline()
```

```
# Runs full data->train->cv->forecast pipeline.
result = pipeline.run_full_pipeline(download=False)
result['summary']
```

```
# Accuracy report for model comparison (lower is better).
cv_summary = result['cv_summary']
cv_summary
```

```
best_model = cv_summary.iloc[0]['model']
best_smape = cv_summary.iloc[0]['smape']
best_wape = cv_summary.iloc[0]['wape']
{'best_model': best_model, 'smape': float(best_smape), 'wape': float(best_wape)}
```

```
# Forecast after model is trained and persisted.
preds = pipeline.forecast(horizon=14, ids=['AAPL.US', 'MSFT.US', 'GOOG.US'])
preds.head()
```
